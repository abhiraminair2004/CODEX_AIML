```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

```

```python
df=pd.read_csv("/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv")
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of the data')
plt.show()
```

```python
x=df[['YearsExperience']]
y=df[['Salary']]
```

```python
model=LinearRegression()
model.fit(x,y)
slope=model.coef_[0]
intercept=model.intercept_
```

```python
def linear_regression(x):
    return slope*x+intercept

    
```

```python
plt.scatter(df['YearsExperience'],df['Salary'],label='Data Points')
plt.plot(df['YearsExperience'],[linear_regression(xi) for xi in df['YearsExperience']],color='red', label="Regression")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```

```python
new_x=6
prediction=linear_regression(new_x)
print(f"Prediction for x={new_x}:{prediction}")
```
